<h2>data preprocessing</h2><br>
1. remove missing values<br>
2. data normalization (min-max scalling)

```python
#removing
df.dropna(axis=0,inplace=True) 
```

```python
df.isnull().sum()
```

```python
#x- independent
x=df.iloc[:,1:].values
#y- dependent
y=df.iloc[:,0].values
```

```python
x.shape
```

```python
y.shape
```

<h3>formula for min-max scalling:<br>
Xnorm = $\frac{x - minValue}{maxValue - minValue}$
</h3>

since data is images, it is expected all the values to be in the range 0 to 2^n-1. n is 8 bit image, so range is (0, 255)

```python
x.min(), x.max()
```

```python
xnorm= x - 0 / 255 - 0
```

```python
xnorm= x / x.max()
```

```python
xnorm
```

now all the values are in the range of 0 to 1, ie, data is normalized

```python
xnorm.shape
```

```python
y
```

```python
#replacing
ynorm=np.where(y=='female',1,0) #replace with 1 else 0
```

```python
ynorm
```

for normalization purpose, we've given the values in the range (0,1)

```python
#saving xnorm and ynorm in zip using numpy
np.savez('./data/data_100_100_norm.npz',xnorm,ynorm)
```
