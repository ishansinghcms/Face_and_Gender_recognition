<h3>resizing</h3>

<b>step1. read image<br>
step2. convert the image into grayscale<br>
step3. resize to 100 x 100<br>
step4. flatten the image (1 x 10,000)</b>

```python
 def resize_img(path_to_resize):
    img=cv2.imread(path_to_resize)
    #conversion
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    #resize
    size=gray.shape[0]
    if size>100:
        gray_re=cv2.resize(gray,(100,100),cv2.INTER_AREA) # shrink
    else:
        gray_re=cv2.resize(gray,(100,100),cv2.INTER_CUBIC) # enlarge
    #flatten
    return gray_re.flatten()
```
